## 시각화 패키지 Matplotlib
###  pylab 서브패키지

```python
import matplotlib as mpl
import matplotlib.pylab as plt
%matplotlib inline
```

### 라인 플롯

```python
plt.title('plot')
plt.plot([1,4,9,16])
plt.show()
```

```python
plt.title('x ticks')
plt.plot([10,20,30,40],[1,4,9,16])
plt.show()
```

### 한글폰트 사용

```python
set(sorted([
    f.name for f in mpl.font_manager.fontManager.ttflist
]))
```

```python
mpl.rc('font', family='Apple SD Gothic Neo')
mpl.rc('axes', unicode_minus=False)
```

```python
plt.title('한글 제목')
plt.plot([10,20,30,40],[1,4,9,16])
plt.xlabel('엑스')
plt.ylabel('와이')
plt.show()
```

```python
font = {'family': 'Apple SD Gothic Neo', 'size': 24, 'color': 'orange'}
plt.title('한글 제목', fontdict=font)
plt.plot([10,20,30,40],[1,4,9,16])
plt.xlabel('엑스')
plt.ylabel('와이')
plt.show()
```

### 스타일 지정

```python
plt.plot([10,20,30,40],[1,4,9,16], 'rs--')
plt.show()
```

###  그림 범위 지정

```python
plt.plot(
    [10, 20, 30, 40],
    [1 ,4, 9, 16],
    c='b',
    lw=5,
    ls='--',
    marker='o',
    ms=15,
    mec='g',
    mew=5,
    mfc='r'
)
plt.xlim(0, 50)
plt.ylim(-10, 30)
plt.show()
```

### 틱 설정

```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
plt.yticks([-1, 0, 1])
plt.show()
```

```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.plot(X, C)
plt.xticks(
    [-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
    [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$']
)
plt.yticks([-1, 0, 1], ["Low", "Zero", "High"])
plt.show()
```

### 그리드 설정

```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.plot(X, C)
plt.xticks(
    [-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
    [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$']
)
plt.yticks([-1, 0, 1], ["Low", "Zero", "High"])
plt.grid(True)
plt.show()
```

### 여러 개의 선 그리기

```python
t = np.arange(0., 5., 0.2)
plt.plot(t, t, 'r--', t, 0.5 * t**2, 'bs:', t, 0.2 * t**3, 'g^-')
plt.show()
```

### 겹쳐 그리기

```python
plt.plot(
    [1 ,4, 9, 16],
    c='b',
    lw=5,
    ls='--',
    marker='o',
    ms=15,
    mec='g',
    mew=5,
    mfc='r'
)
plt.plot(
    [9, 16, 4, 1],
    c='k',
    lw=3,
    ls=':',
    marker='s',
    ms=10,
    mec='m',
    mew=5,
    mfc='c'
)
plt.show()
```
