## 12. 위치 정보형

### 12.1. 한국 측지계를 세계 측지계로 변환

#### 한국 측지계를 세계 측지계로 변환

```python
import pyproj

def convert_to_contunous(x):
    x_min = (x * 100 - int(x * 100)) * 100
    x_sec = (x - int(x) - x_min / 10000) * 100
    return int(x) + x_sec / 60 + x_min / 60 / 60

customer_tb['home_latitude'] = customer_tb['home_latitude'].apply(
    lambda x: convert_to_contunous(x)
)
customer_tb['home_longitude'] = customer_tb['home_longitude'].apply(
    lambda x: convert_to_contunous(x)
)

epsg_world = pyproj.Proj('+init=EPSG:4326')

epsg_korea = pyproj.Proj('+init=EPSG:4301')

home_position = customer_tb[['home_longitude', 'home_latitude']].apply(
    lambda x: pyproj.transform(epsg_korea, epsg_world, x[0], x[1]),
    axis=1
)

customer_tb['home_longitude'] = [x[0] for x in home_position]
customer_tb['home_latitude'] = [x[1] for x in home_position]
customer_tb[['home_latitude', 'home_longitude']]
```

```python
def hubeny(lon1, lat1, lon2, lat2, a=6378137, b=6356752.314245):
    e2 = (a ** 2 - b ** 2) / a ** 2
    (lon1, lat1, lon2, lat2) = \
        [x * (2 * math.pi) / 360 for x in (lon1, lat1, lon2, lat2)]
    w = 1 - e2 * math.sin((lat1 + lat2) / 2) ** 2
    c2 = math.cos((lat1 + lat2) / 2) ** 2
    return math.sqrt(
        (b ** 2 / w ** 3) * (lat1 - lat2) ** 2 +
        (a ** 2 / w) * c2 * (lon1 - lon2) ** 2
    )

home_and_hotel_points.apply(
    lambda x: hubeny(x[0], x[1], x[2], x[3]),
    axis=1
)
```
